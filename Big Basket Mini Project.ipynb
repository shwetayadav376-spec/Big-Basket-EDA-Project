```python
import numpy as np
def handle_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

   
    mean_value = df[column].mean()
    df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound),
                          mean_value,
                          df[column])
handle_outliers('sale_price')
handle_outliers('market_price')
handle_outliers('rating')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Category-wise product count
plt.figure()
colors = ['skyblue','lightgreen','pink','purple','yellow','orange','grey','cyan','olive','brown','blue']
df['category'].value_counts().plot(kind='bar',color = colors, edgecolor = 'black', linewidth = 0.5)
plt.title("CATEGORY-WISE PRODUCT DISTRIBUTION")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Distribution of Sale Price
plt.figure()
color = 'orange'
plt.hist(df['sale_price'], color = color, edgecolor = 'black', linewidth = 0.1)
plt.title("DISTRIBUTION OF SALE PRICE")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3. Boxplot for detecting outliers in price
plt.figure()
color = 'purple'
sns.boxplot(x=df['sale_price'], color = color)
plt.title("BOXPLOT OF SALE PRICE")
plt.show()
```

```python
# 4. Scatter plot for Price vs Rating
plt.figure()
color = ['lightgreen']
plt.scatter(df['sale_price'], df['rating'], color = color, edgecolor = 'black', linewidth = 0.1)
plt.title("SALE PRICE vs RATING")
plt.xlabel("Sale Price")
plt.ylabel("Rating")
plt.show()
```
